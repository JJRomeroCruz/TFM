```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 5)
y1 = np.exp(-0.5*x) + 0.3*np.exp(-1.7*x)
y2 = 0.5*np.exp(-0.5*x) + 1*np.exp(-1.7*x)

plt.plot(x, y1, label = r'$D(p^c(t), T_b)$')
plt.plot(x, y2, label = r'$D(p^h(t), T_b)$')
plt.xlabel('t')
plt.ylabel(r'D($p_i(t)$, $T_b$)')
plt.legend()
```
